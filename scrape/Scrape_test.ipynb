```python
from azure.storage.blob import BlobServiceClient, BlobClient, BlobSasPermissions, generate_blob_sas
import pandas as pd
import requests 
from bs4 import BeautifulSoup, NavigableString
import sys
import re
import random
import time

```

```python
def get_main_tags(url):
    '''
    A function to scrape quotes by type of quote. 
    Arguments: 
        tag --> the type or genre of quote
        page_num --> the page that is to be scraped
    returns: 
        all_quotes --> all quotes under the given tag 
    '''
    compile_1 = re.compile('<.*?>')
    compile_2 = re.compile(r'\s+\d+\.*\d*k\s*')
    all_tags = []
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    popular_tags = soup.find_all(class_='listTagsTwoColumn')
    
    for i in popular_tags: 
        tags = re.sub(compile_1, '', str(i))
        tags = re.sub(compile_2, '', tags)
        clean = tags.replace('Quotes', ',').replace('\n', '').split(',')
        for j in clean:
            all_tags.append(j.strip())
            
    all_tags = [i for i in all_tags if i != '']
        
    return all_tags
```

```python
tags = get_main_tags('https://www.goodreads.com/quotes')
```

```python
def scrape_quotes(tag, page_num=1):
    '''
    A function to scrape quotes by type of quote. 
    Arguments: 
        tag --> the type or genre of quote
        page_num --> the page that is to be scraped
    returns: 
        all_quotes --> all quotes under the given tag 
    '''
    
    old_tag = tag
    tag = tag.replace(" ", "+")
    all_quotes = []

    #handle page unspecified page numbers
    if page_num is None: 
        try: 
            page = requests.get(f"https://www.goodreads.com/quotes/tag/{tag}")
            soup = BeautifulSoup(page.text, 'html.parser')
            page_num = int(soup.find('em', class_='current').text)
        except:
            page_num = 1


    for i in range(1, page_num+1):
        wait_time = random.randint(1, 5)
        time.sleep(wait_time)
        
        try: 
            page = requests.get(f"https://www.goodreads.com/quotes/tag/{tag}?page={i}")
            soup = BeautifulSoup(page.text, 'html.parser')
            print("scraping page", i, 'for', tag)
        except:
            print("Could not connect to site")
            break

        try:
            quote = soup.find(class_="leftContainer")
            quote_list = quote.find_all(class_="quoteDetails")

        except:
            pass

        for quote in quote_list:

            meta_data = []

            try:
                text = quote.find(class_='quoteText').get_text()
                final_quote = text.split('\n')[1]
                meta_data.append(final_quote.strip())
            except:
                pass


            try:
                author = quote.find(class_="authorOrTitle").get_text() 
                autor = author.replace(",", "")
                meta_data.append(author.strip())
            except:
                meta_data.append(None)


            try:
                tags = quote.find(class_="greyText smallText left").get_text()
                tags = [x.strip() for x in tags.split(",")]
                tags = tags[1:]
                meta_data.append(tags)
            except:
                meta_data.append(None)
                
            all_quotes.append(meta_data)
    

    return all_quotes
```

```python
popular_tags = []
for i in tags:
    popular_tags.append(scrape_quotes(i, 100))
```
