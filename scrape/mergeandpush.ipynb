```python
df1 = pd.read_csv('brainy_final.csv', index_col = False)
df1.drop(['page', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df2 = pd.read_csv('goodreadsvf.csv')
df2.quote = df2.quote.replace({'“':'', "”":''}, regex=True)
```

```python
df2['tags'] = df2['tags'].apply(ast.literal_eval) #convert to list type
df2 = df2.explode('tags')
```

```python
new_df = df2.append(df1)
new_df.tags = new_df.tags.replace({'-':' '}, regex=True)
```
